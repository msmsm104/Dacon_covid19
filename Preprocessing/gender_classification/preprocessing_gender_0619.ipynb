```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
train_df = pd.read_csv('../Data/Interim/mfcc/train_mfcc_data(32).csv')
unlabeled_df = pd.read_csv('../Data/Interim/unlabeled/unlabeled_mfcc_data.csv')
test_df = pd.read_csv('../Data/Interim/mfcc/test_mfcc_data(32).csv')
```

```python
train = train_df.drop(columns=['id','age','respiratory_condition','fever_or_muscle_pain', 'covid19'])
```

```python
test = test_df.drop(columns=['id','age','respiratory_condition','fever_or_muscle_pain'])
```

```python
unlabeled = unlabeled_df.drop(columns=['id','age','respiratory_condition','fever_or_muscle_pain'])
```

```python
data = pd.concat([train, test, unlabeled])
```

```python
no_other_df = data[data['gender'] != 'other']
other_df = data[data['gender'] == 'other']
```

```python
no_other_df['gender'] = no_other_df['gender'].map(lambda x : 0 if x == 'male' else 1)
no_other_df
```
