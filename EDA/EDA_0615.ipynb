```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

import warnings

warnings.simplefilter('ignore')
```

```python
f, ax = plt.subplots(32, 1, figsize=(6, 180))

for i in range(1, 32+1):
    sns.distplot(train_32[train_32['covid19'] == 0][f"mfcc_{i}"], ax=ax[i-1], color='green')
    sns.distplot(train_32[train_32['covid19'] == 1][f"mfcc_{i}"], ax=ax[i-1], color='red')



plt.show()
```
